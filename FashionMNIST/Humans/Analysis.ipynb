## Analysis of reaction time

```python
import os
import pandas as pd
import numpy as np
```

```python
base = 'Raw_data_edited/'
files = os.listdir(base)
```

```python
len(files)
```

```python
files
```

```python
reaction_time = np.zeros(38)
reaction_counts = np.zeros(38)
for file in files:
    try:
        data = pd.read_csv(base+file)
    except:
        continue
    for row in data.iterrows():
        stimulus=  row[1]['stimulus']
        image_no = int(stimulus.split('_')[1])
        if np.isnan(row[1]['RSP_pleasant.rt']):
            print('nan')
            continue
        reaction_time[image_no] += float(row[1]['RSP_pleasant.rt'])
        reaction_counts[image_no] += 1
```

```python
reaction_time_avg=  reaction_time/reaction_counts
```

```python
reaction_time_avg
```

```python
search_results_path = '../Search/Fashion Mnist Results - Fmnist_11.csv'
```

```python
search_results = pd.read_csv(search_results_path)
```

```python
search_results.head()
```

```python
search_iterations = np.zeros(38)
count = 0
for row in search_results.iterrows():
    if np.isnan(row[1]['Iteration']):
        search_iterations[count] = 31
    else:
        search_iterations[count] = int(row[1]['Iteration'])
    count += 1
    if(count >37):
        break
    
    
    
        
        
    
    
```

## Noise cealing

```python
reaction_times = np.zeros((11,38))
#reaction_counts = np.zeros(38)
i = 0
for file in files:
    try:
        data = pd.read_csv(base+file)
    except:
        continue
    for row in data.iterrows():
        stimulus=  row[1]['stimulus']
        image_no = int(stimulus.split('_')[1])
        if np.isnan(row[1]['RSP_pleasant.rt']):
            reaction_times[i,image_no] = 30
            continue
        reaction_times[i,image_no] = float(row[1]['RSP_pleasant.rt'])
    i += 1
```

```python
import numpy as np
from scipy.stats import pearsonr

# Example: Simulated reaction time data (stimuli x participants)
data = np.transpose(reaction_times)  # 100 stimuli, 10 participants

# Compute noise ceiling using leave-one-out correlations
noise_ceiling = []
for i in range(data.shape[1]):
    others = np.mean(np.delete(data, i, axis=1), axis=1)
    corr, _ = pearsonr(data[:, i], others)
    noise_ceiling.append(corr)

average_noise_ceiling = np.mean(noise_ceiling)
print(f"Average Noise Ceiling: {average_noise_ceiling:.3f}")
```
